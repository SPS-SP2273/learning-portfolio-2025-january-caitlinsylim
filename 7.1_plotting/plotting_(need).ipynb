### 1 scatter & 1 line

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y1 values',
         c='blue', ls='-.') 
plt.scatter(x, y_2, label='Y2 values',
         c='red')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.20)
plt.legend(loc='upper left') 
#plt.legend(loc=[0, 0.7])
plt.show
```

### With Error Bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = np.array([0, 1, 2, 3, 4, 5, 6])*0.1, np.array([0, 2, 4, 6, 8, 10, 12])*0.1
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

#plt.plot(x, x2, label='x^2',
         #color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

This looks weird and not a smooth curve bcos it is discrete -> we nd to make it continuous 

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 2*np.pi, 1000) # 1000 means plotting at 1000 pts instead of 7, creating smooth curve

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='exp(x)',
         color='red', linestyle='dashed', )
plt.legend(loc='upper right') #legend comes after plt.plot
plt.twinx() #need to put this in btw and not LAST line -> create new y-axis for plots that come after 
plt.plot(x, sin_x, label='sin(x)',
         color='blue')  
plt.legend(loc='upper left')
#plt.twinx() -> putting it here means you create a new y-axis now, but no use bcos nth is behind to be plotted
```

### Plotting sin and cos graphs

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 2*np.pi, 1000) # 1000 means plotting at 1000 pts instead of 7, creating smooth curve

x2 = x**2             
sin_x = np.sin(x)
cos_x = np.cos(x)

plt.plot(x, cos_x, label='cos(x)',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()
plt.twinx() #share same x axis, but y axis values differ 
```
